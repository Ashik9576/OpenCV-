# **Draw geometric shapes on images using Python OpenCV**

```python
color_image=cv2.line(img,(0,0),(400,400),(0,0,400),6)
color_image=cv2.arrowedLine(img,(0,255),(255,255),(34,55,50),6)
cv2_imshow(color_image)
```

```python
color_image=cv2.rectangle(img,(50,50),(350,350),(134,55,50),6) #instead of thickness value 6 if we provide -1 the it will fill the rectangle with same color 
cv2_imshow(color_image)
```

```python
color_image=cv2.circle(img,(200,200),150,(0,155,0),-1) # due to thikness value is -1
cv2_imshow(color_image)
```

```python
font=cv2.FONT_HERSHEY_COMPLEX
color_image=cv2.putText(img,'MODI',(20,200),font,3,(155,155,0),6,cv2.LINE_4) 
cv2_imshow(color_image)
```
